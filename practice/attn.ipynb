```python
import torch
import torch.nn as nn
```

```python
class BahdanauAttention(nn.Module):
    def __init__(self):
        super().__init__()
        self.W = nn.Linear(2*128, 64)
        self.V = nn.Linear(64,1)
        self.softmax = nn.Softmax(dim=0)

    def forward(self, s0, encoder_all):
        encoder = torch.cat(encoder_all, dim=0)
        decoder = s0.unsqueeze(0).repeat(encoder.shape[0], 1)
        print(encoder.shape[0],decoder.shape)
        input_matrix = torch.cat((decoder, encoder), dim=1)
        print(input_matrix.shape)
        x = torch.tanh(self.W(input_matrix))
        print(x.shape)
        e = self.V(x)
        print(e.shape)
        alpha = self.softmax(e)
        c = alpha.T @ encoder
        return c
```

```python
s0 = torch.randn(128)
h_all = [torch.rand(1,128) for i in range(4)]
attn = BahdanauAttention()
```

```python
out = attn.forward(s0, h_all)
```

```python
class LuongAttention(nn.Module):
    def __init__(self):
        super().__init__()
        self.softmax = nn.Softmax(dim=0)

    def forward(self, s1, h_all):
        h = torch.cat(h_all, dim=0)
        e = h @ s1.T
        print(e.shape)
        alpha = self.softmax(e)
        c = alpha.T @ h
        return c
```

```python
attn1 = LuongAttention()
s0 = torch.randn(1,128)
h_all = [torch.rand(1,128) for i in range(4)]
out = attn1.forward(s0, h_all)
out.shape
```
